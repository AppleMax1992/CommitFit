```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
import numpy as np
# Example data
x_values = [0, 1, 2, 3, 4]  # Number of intervals
x_ticks = [5, 10, 15, 20, 50]  # X-axis tick values

# The sets of y-values is collected from E-2 folders

y1 = [40.59, 41.69, 44.08, 37.86, 35.87]   # Accuracy values for bert
y2 = [42.58, 55.00, 53.45, 57.20, 63.03]  # Accuracy values for ST
y3 = [52.81, 68.42, 69.13, 70.49, 75.03]   # Accuracy values for BooCC_{mge}
y4 = [49.66, 69.34, 69.54, 71.42, 74.54] # Accuracy values for BooCC_{mge+cc}
y5 = [46.90, 62.68, 67.47, 71.42, 75.93] # Accuracy values for BooCC_{mge}^{RoBERTa}
y6 = [51.29, 63.93, 68.50, 71.25, 75.81] # Accuracy values for BooCC_{mge+cc}^{RoBERTa}

```

```python
# Plotting the curves with different line styles and markers
plt.plot(x_values, y1, linestyle='--', marker='s', label='BERT')
plt.plot(x_values, y2, linestyle='-.', marker='d', label='ST')
plt.plot(x_values, y3, linestyle='-', marker='o', label='$\mathbf{BooCC_{mge}}$')

plt.plot(x_values, y4, linestyle=':', marker='^', label='$\mathbf{BooCC_{mge+cc}}$')
plt.plot(x_values, y5, linestyle='-', marker='', label='$\mathbf{BooCC_{mge}^{RoBERTa}}$')
plt.plot(x_values, y6, linestyle=':', marker='*', label='$\mathbf{BooCC_{mge+cc}}$')


font = FontProperties()
font.set_weight('bold')

plt.xlabel('Training samples')
plt.ylabel('Accuracy')
plt.title('Dataset II (3-way)')
plt.grid(True)
plt.legend(loc="lower right",prop=font)

# Set custom x-axis ticks and tick labels
plt.xticks(x_values, x_ticks)

plt.show()
```
