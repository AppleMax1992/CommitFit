```python
import matplotlib.pyplot as plt
import numpy as np
# Example data
x_values = [0, 1, 2, 3, 4]  # Number of intervals
x_ticks = [5, 10, 15, 20, 50]  # X-axis tick values

# The sets of y-values is collected from E-2 folders
y1 = [49.83, 66.63, 69.43, 70.49, 73.94]   # Accuracy values for curve 1
y2 = [53.09, 66.87, 70.06, 70.62, 73.88]   # Accuracy values for curve 2
y3 = [43.92, 65.69, 67.01, 71.31, 76.36] # Accuracy values for curve 3
y4 = [43.92, 65.69, 67.01, 71.31, 76.36]  # Accuracy values for curve 4
```

```python
# Plotting the curves with different line styles and markers
plt.plot(x_values, y1, linestyle='-', marker='o', label='$BooCC_{mge}$')
plt.plot(x_values, y2, linestyle='--', marker='s', label='$BooCC_{mge+cc}$')
plt.plot(x_values, y3, linestyle='-.', marker='^', label='$BooCC_{mge}^{RoBERTa}$')
plt.plot(x_values, y4, linestyle=':', marker='d', label='$BooCC_{mge+cc}^{RoBERTa}$')

plt.xlabel('Training samples')
plt.ylabel('Accuracy')
plt.title('Dataset II (3-way)')
plt.grid(True)
plt.legend()

# Set custom x-axis ticks and tick labels
plt.xticks(x_values, x_ticks)

plt.show()
```
