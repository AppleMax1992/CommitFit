```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
import numpy as np
# Example data
x_values = [0, 1, 2, 3, 4]  # Number of intervals
x_ticks = [5, 10, 15, 20, 50]  # X-axis tick values

# The sets of y-values is collected from E-2 folders

y1 = [31.47, 33.10, 39.71, 33.97, 41.09]   # Accuracy values for BERT
y2 = [43.35, 51.42, 53.45, 57.20, 63.03]  # Accuracy values for ST
y3 = [49.83, 66.63, 69.43, 70.49, 73.94]   # Accuracy values for $BooCC_{mge}$
y4 = [43.92, 65.69, 67.01, 71.31, 76.36] # Accuracy values for $BooCC_{mge}^{RoBERTa}$

```

```python
# Plotting the curves with different line styles and markers
plt.plot(x_values, y1, linestyle='--', marker='s', label='BERT')
plt.plot(x_values, y2, linestyle=':', marker='d', label='ST')
plt.plot(x_values, y3, linestyle='-', marker='o', label='$\mathbf{BooCC_{mge}}$')
plt.plot(x_values, y4, linestyle='-.', marker='^', label='$\mathbf{BooCC_{mge}^{RoBERTa}}$')


font = FontProperties()
font.set_weight('bold')

plt.xlabel('Training samples')
plt.ylabel('Accuracy')
plt.title('Dataset II (3-way)')
plt.grid(True)
plt.legend(loc="lower right",prop=font)

# Set custom x-axis ticks and tick labels
plt.xticks(x_values, x_ticks)

plt.show()
```
