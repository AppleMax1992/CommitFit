```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.font_manager import FontProperties
# X-values
x_values = ['CR-ds', 'Transformer', 'E-SPI', 'BERT', 'ST', '$\mathbf{BooCC_{mge}}$', '$\mathbf{BooCC_{mge}^{RoBERTa}}$']

# Y-values (random data)
y_values = [84.1, 79.43, 87.66, 86.61, 85.59, 90.01, 90.54]  # Generates 5 random integers between 1 and 10
print(len(x_values))
print(len(y_values))
# Set colors for the bars
colors = ['#ECF1FF','#FFF2CC','#B7DDE8', '#B6DAA8', '#FEE599', '#789440','#D9958F']

# Adjust figure size
plt.figure(figsize=(12, 8))  # Set a larger figure size (width: 10 inches, height: 6 inches)
for i, score in enumerate(y_values):
    plt.text(i, score, str(score), ha='center', va='bottom', fontweight = 'bold', fontsize=12)
# Create a bar chart with custom colors
bars = plt.bar(x_values, y_values, color=colors, width=0.5, zorder=3)

# Set the font properties for the x-axis labels
font = FontProperties()
font.set_weight('bold')
font.set_size(12)
# Apply the bold font to the x-axis tick labels
plt.xticks(x_values, fontproperties=font)

# Add labels and title
# plt.xlabel('X-values')
plt.ylabel('Accuracy')
plt.title('Acc on Dataset I (2-way)', fontsize=25)

# Set the Y-axis limits
plt.ylim(bottom=80)  # Set the starting value of the Y-axis to 0

# Set the Y-axis tick interval
plt.yticks(np.arange(70, 100, 2))

# Create a legend
legend_labels = ['CR-ds', 'Transformer', 'E-SPI', 'BERT', 'ST', '$\mathbf{BooCC_{mge}}$', '$\mathbf{BooCC_{mge}^{RoBERTa}}$']
plt.legend(bars, legend_labels,prop=font)

# Add horizontal gridlines
plt.grid(axis='y', linewidth=0.5, zorder=1)  # Set zorder=1 to place the gridlines behind

# Display the chart
plt.show()
```
