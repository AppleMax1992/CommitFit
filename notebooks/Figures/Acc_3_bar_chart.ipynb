```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
import numpy as np

x_values = ['$\mathbf{DNN@BERT}$','$\mathbf{DNN@BERT+Fix\_cc}$', 'BERT', 'ST','$\mathbf{BooCC_{mge}}$', '$\mathbf{BooCC_{mge}^{RoBERTa}}$','$\mathbf{BooCC_{mge+cc}}$','$\mathbf{BooCC_{mge+cc}^{RoBERTa}}$']

# Y-values (random data)
y_values = [77.15, 78.89, 80.48,71.00, 81.97, 78.99, 81.78, 82.89]  # Generates 5 random integers between 1 and 10

# Set colors for the bars
colors = ['#ECF1FF','#B7DDE8', '#B6DAA8', '#FEE599', '#789440','#D9958F']

# Adjust figure size
plt.figure(figsize=(14, 8))  # Set a larger figure size (width: 10 inches, height: 6 inches)

# Create a bar chart with custom colors
bars = plt.bar(x_values, y_values, color=colors, width=0.45, zorder=3)

# Set the font properties for the x-axis labels
font = FontProperties()
font.set_weight('bold')
# font.set_size(12)
# Apply the bold font to the x-axis tick labels
plt.xticks(x_values, fontproperties=font)
# Add labels and title
# plt.xlabel('X-values')
plt.ylabel('Accuracy')
plt.title('Acc on Dataset II (3-way)', fontsize=20)
# Set the Y-axis limits
plt.ylim(bottom=80)  # Set the starting value of the Y-axis to 0

# Set the Y-axis tick interval
plt.yticks(np.arange(70, 90, 2))

# Add scores on top of each bar
for i, score in enumerate(y_values):
    plt.text(i, score, str(score), ha='center', va='bottom', fontweight = 'bold',fontsize=12)

# Create a legend
legend_labels = ['$\mathbf{DNN@BERT}$','$\mathbf{DNN@BERT+Fix\_cc}$', 'BERT', 'ST','$\mathbf{BooCC_{mge}}$', '$\mathbf{BooCC_{mge}^{RoBERTa}}$','$\mathbf{BooCC_{mge+cc}}$','$\mathbf{BooCC_{mge+cc}^{RoBERTa}}$']

plt.legend(bars, legend_labels,prop=font)

# Add horizontal gridlines
plt.grid(axis='y', linewidth=0.5, zorder=1)  # Set zorder=1 to place the gridlines behind

# Display the chart
plt.show()
```
