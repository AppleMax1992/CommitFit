```python
import matplotlib.pyplot as plt
import numpy as np

x_values = ['DNN@BERT+Fix_cc', '$BooCC_{mge}$', '$BooCC_{mge+cc}$', '$BooCC_{mge}^{RoBERTa}$', '$BooCC_{mge+cc}^{RoBERTa}$']

# Y-values (random data)
y_values = [78.89, 79.74, 79.74, 82.34, 82.34]  # Generates 5 random integers between 1 and 10

# Set colors for the bars
colors = ['yellow', 'blue', 'orange', 'green','red']

# Adjust figure size
plt.figure(figsize=(10, 6))  # Set a larger figure size (width: 10 inches, height: 6 inches)

# Create a bar chart with custom colors
bars = plt.bar(x_values, y_values, color=colors, width=0.5, zorder=3)


# Add labels and title
# plt.xlabel('X-values')
plt.ylabel('Accuracy')
plt.title('Acc on Dataset II (3-way)')
# Set the Y-axis limits
plt.ylim(bottom=80)  # Set the starting value of the Y-axis to 0

# Set the Y-axis tick interval
plt.yticks(np.arange(70, 90, 2))

# Create a legend
legend_labels = ['DNN@BERT+Fix_cc', '$BooCC_{mge}$', '$BooCC_{mge+cc}$', '$BooCC_{mge}^{RoBERTa}$', '$BooCC_{mge+cc}^{RoBERTa}$']

plt.legend(bars, legend_labels)

# Add horizontal gridlines
plt.grid(axis='y', linewidth=0.5, zorder=1)  # Set zorder=1 to place the gridlines behind

# Display the chart
plt.show()
```
