```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
import numpy as np
# Example data
x_values = [0, 1, 2, 3, 4]  # Number of intervals
x_ticks = [5, 10, 15, 20, 50]  # X-axis tick values

# The sets of y-values are collected from E-2 folders
y1 = [62.76, 63.24, 63.24, 64.63, 65.14]   # Accuracy values for bert
y2 = [68.99, 71.18, 74.77, 76.17, 78.33]  # Accuracy values for ST
y3 = [70.09, 74.12, 76.33, 76.16, 80.47]   # Accuracy values for BooCC_{mge}
y4 = [70.16, 73.30, 76.37, 76.83, 80.84] # Accuracy values for BooCC_{mge+cc}
y5 = [66.54, 70.55, 75.87, 77.86, 81.35] # Accuracy values for BooCC_{mge}^{RoBERTa}
y6 = [67.82, 69.71, 76.58, 78.26, 81.14] # Accuracy values for BooCC_{mge+cc}^{RoBERTa}
```

```python
# Adjust figure size
plt.figure(figsize=(10, 6))  # Set a larger figure size (width: 10 inches, height: 6 inches)
# Plotting the curves with different line styles and markers
plt.plot(x_values, y1, linestyle='--', marker='s', label='BERT')
plt.plot(x_values, y2, linestyle='-.', marker='d', label='ST')
plt.plot(x_values, y3, linestyle='-', marker='o', label='$\mathbf{BooCC_{mge}}$')

plt.plot(x_values, y4, linestyle=':', marker='^', label='$\mathbf{BooCC_{mge+cc}}$')
plt.plot(x_values, y5, linestyle='-', marker='', label='$\mathbf{BooCC_{mge}^{RoBERTa}}$')
plt.plot(x_values, y6, linestyle=':', marker='*', label='$\mathbf{BooCC_{mge+cc}}$')

font = FontProperties()
font.set_weight('bold')
# font.set_size(12)

plt.xlabel('Training samples')
plt.ylabel('Accuracy')
plt.title('Dataset I (2-way)')
plt.grid(True)
plt.legend(loc="lower right",prop=font)

# Set custom x-axis ticks and tick labels
plt.xticks(x_values, x_ticks)
# plt.yticks(np.arange(65, 85, 0.5))
plt.show()
```
