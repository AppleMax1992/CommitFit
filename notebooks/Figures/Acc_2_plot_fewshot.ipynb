```python
import matplotlib.pyplot as plt
import numpy as np
# Example data
x_values = [0, 1, 2, 3, 4]  # Number of intervals
x_ticks = [5, 10, 15, 20, 50]  # X-axis tick values

# The sets of y-values are collected from E-2 folders
y1 = [68.44, 73.05, 75.90, 76.46, 79.81]   # Accuracy values for curve 1
y2 = [68.89, 72.82, 75.94, 76.76, 80.01]   # Accuracy values for curve 2
y3 = [64.60, 70.17, 73.71, 76.63, 79.27] # Accuracy values for curve 3
y4 = [65.27, 70.31, 73.86, 76.77, 79.29]  # Accuracy values for curve 4
```

```python
# Plotting the curves with different line styles and markers
plt.plot(x_values, y1, linestyle='-', marker='o', label='$BooCC_{mge}$')
plt.plot(x_values, y2, linestyle='--', marker='s', label='$BooCC_{mge+cc}$')
plt.plot(x_values, y3, linestyle='-.', marker='^', label='$BooCC_{mge}^{RoBERTa}$')
plt.plot(x_values, y4, linestyle=':', marker='d', label='$BooCC_{mge+cc}^{RoBERTa}$')

plt.xlabel('Training samples')
plt.ylabel('Accuracy')
plt.title('Dataset I (2-way)')
plt.grid(True)
plt.legend(loc="lower right")

# Set custom x-axis ticks and tick labels
plt.xticks(x_values, x_ticks)
# plt.yticks(np.arange(65, 85, 0.5))
plt.show()
```
