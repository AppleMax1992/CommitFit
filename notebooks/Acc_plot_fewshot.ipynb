```python
import matplotlib.pyplot as plt
import numpy as np
# Example data
x_values = [0, 1, 2, 3, 4, 5, 6]  # Number of intervals
x_ticks = [0, 5, 10, 15, 20, 50, 100]  # X-axis tick values

# The sets of y-values is collected from E-2 folders
y1 = [0.85, 0.92, 0.95, 0.97, 0.98, 0.95, 0.9]   # Accuracy values for curve 1
y2 = [0.75, 0.85, 0.9, 0.94, 0.96, 0.92, 0.88]   # Accuracy values for curve 2
y3 = [0.95, 0.97, 0.98, 0.99, 0.995, 0.99, 0.98] # Accuracy values for curve 3
y4 = [0.80, 0.88, 0.92, 0.95, 0.97, 0.96, 0.94]  # Accuracy values for curve 4
```

```python
# Plotting the curves with different line styles and markers
plt.plot(x_values, y1, linestyle='-', marker='o', label='$BooCC_{mge}$')
plt.plot(x_values, y2, linestyle='--', marker='s', label='$BooCC_{mge+cc}$')
plt.plot(x_values, y3, linestyle='-.', marker='^', label='$BooCC_{mge}^{RoBERTa}$')
plt.plot(x_values, y4, linestyle=':', marker='d', label='$BooCC_{mge+cc}^{RoBERTa}$')

plt.xlabel('Training samples')
plt.ylabel('Accuracy')
plt.title('Dataset I (2-way)')
plt.grid(True)
plt.legend()

# Set custom x-axis ticks and tick labels
plt.xticks(x_values, x_ticks)

plt.show()
```
